# Exploratory Data Analysis

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the sample Iris dataset
df = pd.read_csv('/home/bhargavi/us-videos.csv')

# Display the first few rows of the dataframe
print(df.head())
```

```python
df.describe()
```

```python
# Visualize distributions of numerical variables
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
numerical_cols
```

```python
# Histograms for numerical features
df[numerical_cols].hist(figsize=(15, 10), bins=20)
plt.suptitle('Distributions of Numerical Features')
plt.show()
```

```python
# Box plots to identify outliers
plt.figure(figsize=(15, 10))
num_plots = len(numerical_cols)
rows = (num_plots // 2) + (num_plots % 2)
for i, col in enumerate(numerical_cols):
    plt.subplot(rows, 2, i + 1)
    sns.boxplot(y=df[col])
    plt.title(f'Box plot of {col}')
plt.tight_layout()
plt.show()
```

```python
# Visualize distributions of categorical variables
categorical_cols = df.select_dtypes(include=['object']).columns

for col in categorical_cols:
    plt.figure(figsize=(10, 5))
    df[col].value_counts().plot(kind='bar')
    plt.title(f'Distribution of {col}')
    plt.ylabel('Count')
    plt.xticks(rotation=90)
    plt.show()
```

```python
# Visualize datetime variables
datetime_cols = df.select_dtypes(include=['datetime64']).columns

for col in datetime_cols:
    plt.figure(figsize=(10, 5))
    df[col].value_counts().sort_index().plot(kind='line')
    plt.title(f'Time Series of {col}')
    plt.ylabel('Count')
    plt.xticks(rotation=90)
    plt.show()
```

```python
# Correlation matrix for numerical features
if not numerical_cols.empty:
    corr_matrix = df[numerical_cols].corr()
    plt.figure(figsize=(16, 6))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
    plt.title('Correlation Matrix')
    plt.show()
```

```python
sns.scatterplot(data=df,x='views',y='likes')
plt.title('views vs likes')
plt.xlabel('views')
plt.ylabel('likes')
plt.show()
```
